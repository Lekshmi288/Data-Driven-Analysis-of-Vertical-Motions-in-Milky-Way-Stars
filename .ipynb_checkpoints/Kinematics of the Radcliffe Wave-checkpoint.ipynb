```python
import matplotlib.pyplot as plt
from astropy.coordinates import ICRS, Galactic, GalacticLSR, Galactocentric
import astropy.units as u
import astropy.coordinates as apycoords
import pandas as pd
from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
import seaborn as sns
import statistics as stat
import numpy as np
from matplotlib.figure import figaspect
#import plotly
import scipy
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
from astropy.coordinates import SkyCoord
from astropy.coordinates import SkyCoord
from galpy.orbit import Orbit
from galpy.potential import MWPotential2014 
import plotly
from galpy.potential import NFWPotential, HernquistPotential, MiyamotoNagaiPotential
import plotly.express as px
from numpy.polynomial import Polynomial
```

```python

df1 = pd.read_csv('Best_Fit_Wave_Model.csv')
a=df1.x.values
b_c=df1.y.values
c=df1.z.values
```

```python
df1 = pd.read_csv('Giants_traced_radcliffewave_median.csv')
y1=df1.median_y.values
vz1=df1.median_vz.values
err1=df1.err.values

df2 = pd.read_csv('OB_traced_radcliffewave_median.csv')
y2=df2.median_y.values
vz2=df2.median_vz.values
err2=df2.err.values

df3 = pd.read_csv('UMS_traced_radcliffewave_median.csv')
y3=df3.median_y.values
vz3=df3.median_vz.values
err3=df3.err.values

df4 = pd.read_csv('OC_traced_radcliffewave_median.csv')
y4=df4.median_y.values
vz4=df4.median_vz.values
err4=df4.err.values
```

```python

w, h = figaspect(1/2)

fig, (ax1, ax3, ax5, ax7) = plt.subplots(nrows=4, ncols=1, dpi=100, figsize=(8,10))

fig.set_tight_layout(False)

fig.subplots_adjust(hspace=0, wspace=0)
color = 'tab:red'


ax1.errorbar(y1/1000,vz1,yerr=err1,ecolor='black', color=color,label='Giants',ms=5,
capsize=3,elinewidth=2, markeredgewidth=2,fmt='s')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylabel('$\widetilde{V^{\prime}_{z}}$ (km/s)', color=color)
ax1.set_ylim(-7,7)
ax1.set_yticks(np.arange(-6,6.5,3))
ax1.axhline(y=0, color='gray', linestyle='-')


ax3.errorbar(y3/1000,vz3,yerr=err3,ecolor='black', color=color,label='UMS',ms=5,
capsize=3,elinewidth=2, markeredgewidth=2,fmt='s')
ax3.tick_params(axis='y', labelcolor=color)
ax3.set_ylabel('$\widetilde{V^{\prime}_{z}}$ (km/s)', color=color)
ax3.set_ylim(-5,5)
ax3.set_yticks(np.arange(-4,5,2))
ax3.axhline(y=0, color='gray', linestyle='-')


ax5.errorbar(y4/1000,vz4,yerr=err4,ecolor='black', color=color,label='Open Clusters',ms=5,
capsize=3,elinewidth=2, markeredgewidth=2,fmt='s')
ax5.tick_params(axis='y', labelcolor=color)
ax5.set_ylabel('$\widetilde{V^{\prime}_{z}}$ (km/s)', color=color)
ax5.set_ylim(-12,12)
ax5.set_yticks(np.arange(-10,11,5))
ax5.axhline(y=0, color='gray', linestyle='-')
               
               
ax7.errorbar(y2/1000,vz2,yerr=err2,ecolor='black', color=color,label='OB',ms=5,
capsize=3,elinewidth=2, markeredgewidth=2,fmt='s')
ax7.tick_params(axis='y', labelcolor=color)
ax7.set_ylabel('$\widetilde{V^{\prime}_{z}}$ (km/s)', color=color)
ax7.set_xlabel('Y(kpc)')
ax7.set_ylim(-12,12)
ax7.set_yticks(np.arange(-10,11,5))
ax7.axhline(y=0, color='gray', linestyle='-')


ax2 = ax1.twinx()  # A second axes that shares the same x-axis to make a dual axis plot 
color = 'tab:blue'
ax2.set_ylabel(' Z (kpc)', color='blue') 
ax2.tick_params(axis='y', labelcolor='blue')
ax2.plot(b_c/1000,c/1000,label='Radcliffe Wave Model',color='blue',lw=4)
ax2.set_ylim(-0.19,0.190)
ax2.legend(fontsize=12,loc=4)

ax4 = ax3.twinx() 
color = 'tab:blue'
ax4.set_ylabel(' Z (kpc)', color='blue') 
ax4.tick_params(axis='y', labelcolor='blue')
ax4.plot(b_c/1000,c/1000,label='Radcliffe Wave Model',color='blue',lw=4)
ax4.set_ylim(-0.190,0.190)
ax4.legend(fontsize=12,loc=4)

ax6 = ax5.twinx() 
color = 'tab:blue'
ax6.set_ylabel(' Z (kpc)', color='blue') 
ax6.tick_params(axis='y', labelcolor='blue')
ax6.plot(b_c/1000,c/1000,label='Radcliffe Wave Model',color='blue',lw=4)
ax6.set_ylim(-0.190,0.190)
ax6.legend(fontsize=12,loc=4)

ax8 = ax7.twinx()  
color = 'tab:blue'
ax8.set_ylabel(' Z (kpc)', color='blue') 
ax8.tick_params(axis='y', labelcolor='blue')
ax8.plot(b_c/1000,c/1000,label='Radcliffe Wave Model',color='blue',lw=4)
ax8.set_ylim(-0.190,0.190)
ax8.legend(fontsize=12,loc=4)

ax1.set_xlim(min(b_c)/1000,max(b_c)/1000)
ax3.set_xlim(min(b_c)/1000,max(b_c)/1000)
ax5.set_xlim(min(b_c)/1000,max(b_c)/1000)
ax7.set_xlim(min(b_c)/1000,max(b_c)/1000)


fig.subplots_adjust(top=0.95)
sns.set_context("talk")
sns.set_style("white")


ax1.text(-650/1000,5.5,'Giant stars',horizontalalignment='center',c='black',fontsize=14,fontweight='bold')
ax3.text(-650/1000,4,'UMS stars',horizontalalignment='center',c='black',fontsize=14,fontweight='bold')
ax5.text(-580/1000,10,'Open Clusters',horizontalalignment='center',c='black',fontsize=14,fontweight='bold')
ax7.text(-670/1000,10,'OB stars',horizontalalignment='center',c='black',fontsize=14,fontweight='bold')

plt.show()

```
