# Rotate the Coordinate system 25 degrees counterclockwise so that the RW is the new y axis

The RW appears as a nearly straight line when projected in the
X âˆ’ Y plane, which makes it convenient to define a coordinate
system aligned with this feature. We rotate the coordinate system
25 degrees anticlockwise about the Z axis, so that the rotated
Y axis (hereafter Yâ€²) lies tangential to and the rotated X axis (hereafter Xâ€²) lies perpendicular to the RW.

```python
# Function for rotating the X-Y plane 25 degrees counterclockwise
def coordinaterotation(x,y):
    xp = ((x) * np.cos(np.radians(-25)) + (y) * np.sin(np.radians(-25)) )
    yp = ( -(x) * np.sin(np.radians(-25)) + (y) * np.cos(np.radians(-25)) )
    return xp,yp
```

We then divide this rotated sample into five panels A, B, C, D and E starting from left with panel C centered on the RW. Each panel is a rectangular region with 1 kpc in width and 3 kpc
in length. We divide Yâ€² into different bins and then calculate median vertical velocities in each bin to look for oscillations beyond the RW.

# Compute Median vertical velocity across each panel along the rotated Y.

```python
import numpy as np
import pandas as pd

def compute_moving_median_and_bootstrap(x, y, vz, n_bootstrap, output_prefix, x_start,x_end, width):
    """
    Computes the moving median and bootstrap errors of Vzprime along Y, within each X panel,
    and saves each region's result to a separate CSV file.

    Parameters:
    - x (np.array): Array of X coordinates (in kpc).
    - y (np.array): Array of Y coordinates (in pc).
    - vz (np.array): Array of Vzprime (vertical velocity) values.
    - n_bootstrap (int): Number of bootstrap resamples.
    - output_prefix (str): Prefix for the output CSV file name (e.g., 'OB').
    - x_start (float): The starting value of X for the first panel.
    - x_end (float): The ending value of X for the first panel.
    - width (float): Width of each X panel (default is 1 kpc).
    
    Returns:
    - None: Saves each X panel's results to a separate CSV file with region-based names.
    """
    # Define Y bins
    bin_min, bin_max = -3,3.3
    bin_width = (bin_max - bin_min) / 21 
    y_bins = np.arange(bin_min, bin_max, bin_width)  # Create overlapping bins


    # Loop over X panels
    x_panels = np.arange(x_start,x_end, width)
    
    for panel_index, x_min in enumerate(x_panels):
        x_max = x_min + width
        
        # Filter the stars that fall within the current X panel range
        mask = (x >= x_min) & (x < x_max)
        filtered_y = y[mask]
        filtered_vz = vz[mask]

        # Initialize lists to store results for the current panel
        results = []

        # We will calculate a simple moving median where we find the median vz values for overlapping bins. i.e 
        # between 0th bin and 2 nd bin. Then between first bin and third bin. Second and fourth bin and so on..
        
        for i in range(0, len(y_bins) - 2):
            vz_array = []

            # Collect Vzprime values within the current Y bin range
            for j in range(len(filtered_y)):
                if y_bins[i] <= filtered_y[j] <= y_bins[i+2]:
                    vz_array.append(filtered_vz[j])
          

            if vz_array:  
                # Compute the median of Vzprime for the current bin
                median_vz = np.nanmedian(vz_array)
                # Compute the midpoint of the Y bin
                bin_center_y = np.mean([y_bins[i], y_bins[i+2]])

                # Perform bootstrap resampling to calculate error
                bootstrap_samples = [np.nanmedian(np.random.choice(vz_array, len(vz_array), replace=True)) 
                                     for _ in range(n_bootstrap)]
                error_med = np.std(bootstrap_samples)
            else:
                # If no data points in the Y bin, append NaN
                median_vz = np.nan
                bin_center_y = np.nan
                error_med = np.nan

            # Store the result for the current X panel and Y bin
            results.append([bin_center_y, median_vz, error_med])

        # Create DataFrame with the results for the current panel
        output_df = pd.DataFrame(results, columns=['y', 'vz', 'error'])

        # Generate file name based on region (A, B, C, ...)
        region_name = chr(65 + panel_index)  # Convert panel index to letter (A = 0, B = 1, etc.)
        output_file = f"{output_prefix}_Region_{region_name}.csv"
        
        # Save the results to the CSV file
        output_df.to_csv(output_file, index=False)

    return output_df
```

```python
# Load the Radcliffe wave model
df1 = pd.read_csv('Best_Fit_Wave_Model.csv')
a=df1.x.values
b_c=df1.y.values
c=df1.z.values

# rotated coordinates for the Radcliffe wave
ap=coordinaterotation(a,b_c)[0]
bp=coordinaterotation(a,b_c)[1]
```

# Analysis for OB sample

```python
#Load the OB sample data
df = pd.read_csv('Cleaned_Data/OB_Stars_cleaned.csv')
x=df.X.values
y=df.Y.values
z=df.Z.values
v_z=df.Vzprime.values
R=df.R.values

# Get the rotated coordinates for the OB sample
xp=coordinaterotation(x,y)[0]
yp=coordinaterotation(x,y)[1]
```

Visualize the Rotated X-Y plane colored as a function of median vertical velocities with the Radcliffe wave :

```python

# Define spatial range and bin size
xmin, xmax = -5, 5
ymin, ymax = -5, 5
rang = [[xmin, xmax], [ymin, ymax]]
binsxy = [int((xmax - xmin) / 0.2), int((ymax - ymin) / 0.2)]

# Compute 2D histogram and median statistic

med_vz = scipy.stats.binned_statistic_2d(
    xp, yp , v_z, statistic='median', range=rang, bins=binsxy
)

# Apply Gaussian smoothing to handle NaNs
sigma = 1.25
U = med_vz.statistic.T.copy()
V = np.nan_to_num(U)  # Replace NaNs with 0 for filtering
VV = ndimage.gaussian_filter(V, sigma=sigma)

# Create weight matrix for normalization
W = np.ones_like(U)
W[np.isnan(U)] = 0
WW = ndimage.gaussian_filter(W, sigma=sigma)

# Compute final smoothed result
np.seterr(divide='ignore', invalid='ignore')  # Suppress divide-by-zero warnings
Z = VV / WW
Z[np.isnan(U)] = np.nan  # Restore NaN values

# Plotting
fig, ax1 = plt.subplots(dpi=150, figsize=(5, 5), sharey=True)
plt.imshow(Z, extent=(xmin, xmax, ymin, ymax), cmap='seismic', origin='lower', vmin=-5, vmax=5)

# Colorbar configuration
bounds = np.arange(-5, 6.0, 2.0)
cbar = plt.colorbar(ticks=bounds, label="Color", fraction=0.046)
cbar.set_label(r"Median $v_{z}$ (km/s)", labelpad=14, y=0.6, rotation=270)

# Labels and title
plt.xlabel(r"$X^{\prime}$ (kpc)")
plt.ylabel(r"$Y^{\prime}$ (kpc)")
plt.title("OB Sample (Rotated)", fontsize=16)

# Seaborn style settings
sns.set_context("talk")
sns.set_style("white")

# Add concentric circles showing different galactocentric radius from 6-12 kpc. Sun is at R=8.2 kpc.
circle_positions = [6, 8, 10, 12]
for r in circle_positions:
    circle = plt.Circle((8.34, 0), r, color='black', fill=False, linestyle='--')
    ax1.add_artist(circle)

# Axis limits and labels
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

# Scatter plot for central marker
plt.scatter(0, 0, marker='+', c='black', lw=3)

# Overlay the Best fit Radcliffe wave model from Alves et al. 2020
plt.plot(ap/1000,bp/1000,c='black',lw=3)

# Show plot
plt.show()
```
